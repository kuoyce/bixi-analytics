````python
# Approach 2: Using chunksize with random sampling from all 3 CSV files
def read_csv_sample_chunks(filepath, sample_size=10000, chunksize=5000):
    """Read CSV in chunks and combine into random sample.
    
    Tracks total rows sampled (not list length) to properly stop at sample_size.
    """
    sampled_rows = []
    total_rows_sampled = 0
    
    for chunk in pd.read_csv(filepath, chunksize=chunksize):
        # Check if we still need more rows
        if total_rows_sampled < sample_size:
            remaining = sample_size - total_rows_sampled
            sample_chunk = chunk.sample(n=min(len(chunk), remaining))
            sampled_rows.append(sample_chunk)
            total_rows_sampled += len(sample_chunk)
        else:
            break
    
    return pd.concat(sampled_rows, ignore_index=True) if sampled_rows else pd.DataFrame()

# Read and sample from all 3 CSV files
print("Reading and sampling 10,000 rows from each CSV file...")
print("=" * 80)

all_samples = []
for i, csv_file in enumerate(csv_files, 1):
    print(f"Processing file {i}/{len(csv_files)}: {csv_file.name}")
    df_sample = read_csv_sample_chunks(csv_file, sample_size=10000, chunksize=5000)
    print(f"  Sampled shape: {df_sample.shape}")
    all_samples.append(df_sample)

# Concatenate all samples
combined_sample_df = pd.concat(all_samples, ignore_index=True)
print("\n" + "=" * 80)
print(f"Combined shape: {combined_sample_df.shape}")
print(f"Total rows sampled: {len(combined_sample_df)}")

# Apply column name mapping
combined_sample_df_en = combined_sample_df.rename(columns=french_to_english)
print(f"\nSample with English headers:")
print(combined_sample_df_en.head(10))
````

## Optimization Strategy & Analysis

### Current Situation
- **Total Data**: ~5 GB across 3 files
- **Main Files**: 2 large files (2.4-2.6 GB each) with 13-14M rows
- **Column Types**: 4 strings (station names, arrondissements), 5 coordinates (floats), 1 timestamp (int64)
- **Memory Expansion**: ~1.7-1.8x file size when loaded (CSV compression is ~60%)

### Key Observations
1. **String columns** are the primary bottleneck (station names repeat frequently - Bikes are reused)
2. **Coordinates** (floats) could use lower precision without significant loss
3. **Timestamps** (milliseconds) are stored as int64 - could be datetime optimized
4. **Data is highly compressible** - CSV is inefficient storage format for this use case

---

### Optimization Options (Ranked by Impact)

#### 1. **Use Parquet Format** ⭐⭐⭐ (Highest Impact)
- **Benefit**: Native columnar compression, 3-5x better compression than CSV
- **File Size**: ~500-800 MB instead of ~5 GB
- **Speed**: 5-10x faster read performance
- **Memory**: Only load columns you need (true columnar benefit)
- **Best for**: One-time conversion + future use

```python
# Convert CSV to Parquet
df = pd.read_csv('large_file.csv')
df.to_parquet('large_file.parquet', compression='snappy', index=False)
# Or use 'gzip' for smaller size (slower) or 'brotli' (fastest, if available)
```

#### 2. **Categorical Encoding for String Columns** ⭐⭐⭐ (Great for Memory)
- **Benefit**: Station names stored as category codes (integers) instead of strings
- **Size Reduction**: 50-70% memory savings on string columns
- **Speed**: Faster filtering/grouping on categorical data
- **Best for**: If working with same dimension values repeatedly

```python
df['STARTSTATIONNAME'] = df['STARTSTATIONNAME'].astype('category')
df['ENDSTATIONNAME'] = df['ENDSTATIONNAME'].astype('category')
# Memory: ~500MB → ~150-200MB for string columns
```

#### 3. **Reduce Float32 Precision** ⭐⭐ (Good for Size)
- **Benefit**: Latitude/Longitude as float32 instead of float64
- **Precision**: float32 ≈ 7 decimal places (enough for 1m precision globally)
- **Size Reduction**: 50% for coordinate columns
- **Best for**: Spatial analysis where millimeter precision isn't needed

```python
df['STARTSTATIONLATITUDE'] = df['STARTSTATIONLATITUDE'].astype('float32')
df['STARTSTATIONLONGITUDE'] = df['STARTSTATIONLONGITUDE'].astype('float32')
# Memory: 4M rows × 8 columns → ~64MB → ~32MB for coordinates
```

#### 4. **Chunked Reading** ⭐⭐ (Good for Processing)
- **Benefit**: Process large files in chunks, don't load entire file to RAM
- **Memory**: Only hold 100k-1M rows at a time
- **Speed**: Slightly slower (I/O overhead) but predictable memory usage
- **Best for**: Aggregations, transformations, statistical analysis

```python
for chunk in pd.read_csv('large_file.csv', chunksize=500000):
    # Process chunk
    result = chunk.groupby('STARTSTATIONNAME').size()
```

#### 5. **DuckDB for Querying** ⭐⭐ (Best for Analytics)
- **Benefit**: SQL queries on CSV/Parquet WITHOUT loading to memory
- **Speed**: 5-100x faster for filtered queries
- **Memory**: Only loads what you query for
- **Best for**: Ad-hoc analysis, aggregations, filtering before loading

```python
import duckdb
result = duckdb.query("""
    SELECT STARTSTATIONNAME, COUNT(*) as trips
    FROM read_parquet('data.parquet')
    GROUP BY STARTSTATIONNAME
    ORDER BY trips DESC
    LIMIT 10
""").to_df()
```

#### 6. **Polars (Alternative to Pandas)** ⭐⭐ (High Performance)
- **Benefit**: Faster processing, lazy evaluation, better memory efficiency
- **Speed**: 5-10x faster than pandas for large datasets
- **Memory**: Better management, lazy loading
- **Best for**: Full replacement of pandas for this workload

```python
import polars as pl
df = pl.read_csv('large_file.csv')
# or for SQL-like queries:
df = pl.scan_csv('large_file.csv').filter(...).select(...).collect()
```

````python
import time
import os
from pathlib import Path
from tabulate import tabulate
import pandas as pd
import warnings
warnings.filterwarnings('ignore', category=UserWarning)

# French to English column mapping (from cell 5)
french_to_english = {
    'STARTSTATIONNAME': 'start_station_name',
    'STARTSTATIONARRONDISSEMENT': 'start_station_district',
    'STARTSTATIONLATITUDE': 'start_station_latitude',
    'STARTSTATIONLONGITUDE': 'start_station_longitude',
    'ENDSTATIONNAME': 'end_station_name',
    'ENDSTATIONARRONDISSEMENT': 'end_station_district',
    'ENDSTATIONLATITUDE': 'end_station_latitude',
    'ENDSTATIONLONGITUDE': 'end_station_longitude',
    'STARTTIMEMS': 'start_time_ms',
    'ENDTIMEMS': 'end_time_ms',
}

print("=" * 80)
print("CONVERTING CSV TO PARQUET WITH OPTIMIZATION")
print("=" * 80)

conversion_results = []

# Define data directory
data_dir = Path("./data")

# Find all CSV files
csv_files = []
for root, dirs, files in os.walk(data_dir):
    for file in files:
        if file.endswith('.csv'):
            csv_files.append(Path(root) / file)

csv_files.sort()

for csv_file in csv_files:
    print(f"\n{'─' * 80}")
    print(f"Processing: {csv_file.name}")
    print(f"{'─' * 80}")
    
    # Create output parquet filename
    parquet_filename = csv_file.stem + '.parquet'
    parquet_file = csv_file.parent / parquet_filename
    
    try:
        start_time = time.time()
        
        # Step 1: Read CSV
        print(f"  1. Reading CSV file...")
        df = pd.read_csv(csv_file)
        read_time = time.time() - start_time
        print(f"     ✓ Read complete in {read_time:.2f}s")
        
        # Step 2: Rename columns to English
        print(f"  2. Renaming columns to English...")
        start_rename = time.time()
        rename_mapping = {col: french_to_english.get(col, col) for col in df.columns}
        df = df.rename(columns=rename_mapping)
        rename_time = time.time() - start_rename
        print(f"     ✓ Columns renamed in {rename_time:.2f}s")
        
        # Step 3: Convert coordinates to float32
        print(f"  3. Converting coordinates to float32...")
        start_float = time.time()
        coord_columns = [col for col in df.columns if 'latitude' in col or 'longitude' in col]
        for col in coord_columns:
            df[col] = df[col].astype('float32')
        float_time = time.time() - start_float
        print(f"     ✓ Float32 conversion complete in {float_time:.2f}s")
        
        # Step 4: Save to Parquet
        print(f"  4. Saving to Parquet (snappy compression)...")
        start_save = time.time()
        df.to_parquet(
            parquet_file, 
            compression='snappy', 
            index=False,
            engine='pyarrow',
            coerce_timestamps='ms',
            use_deprecated_int96_timestamps=False
        )
        save_time = time.time() - start_save
        print(f"     ✓ Parquet saved in {save_time:.2f}s")
        
        # Calculate statistics
        total_time = time.time() - start_time
        original_size_mb = csv_file.stat().st_size / (1024 * 1024)
        parquet_size_mb = parquet_file.stat().st_size / (1024 * 1024)
        compression_ratio = original_size_mb / parquet_size_mb
        memory_before_mb = df.memory_usage(deep=True).sum() / (1024 * 1024)
        
        print(f"\n  📊 Statistics:")
        print(f"     Original CSV: {original_size_mb:.2f} MB")
        print(f"     Parquet file: {parquet_size_mb:.2f} MB")
        print(f"     Compression: {compression_ratio:.2f}x smaller")
        print(f"     Memory usage: {memory_before_mb:.2f} MB")
        print(f"     Total time: {total_time:.2f}s")
        
        conversion_results.append({
            'File': csv_file.name,
            'Status': '✓ Success',
            'Original (MB)': f'{original_size_mb:.2f}',
            'Parquet (MB)': f'{parquet_size_mb:.2f}',
            'Compression': f'{compression_ratio:.2f}x',
            'Time (s)': f'{total_time:.2f}'
        })
        
    except Exception as e:
        error_msg = str(e)[:50]
        print(f"  ✗ Error: {error_msg}")
        conversion_results.append({
            'File': csv_file.name,
            'Status': f'✗ {error_msg}',
            'Original (MB)': '—',
            'Parquet (MB)': '—',
            'Compression': '—',
            'Time (s)': '—'
        })

print(f"\n{'=' * 80}")
print("CONVERSION SUMMARY")
print(f"{'=' * 80}")
print(tabulate(conversion_results, headers='keys', tablefmt='grid'))

# Verify all parquet files were created
print(f"\n{'=' * 80}")
print("VERIFICATION")
print(f"{'=' * 80}")
parquet_files = list(data_dir.rglob('*.parquet'))
print(f"✓ Found {len(parquet_files)} Parquet files:")
for pf in sorted(parquet_files):
    size_mb = pf.stat().st_size / (1024 * 1024)
    print(f"  • {pf.relative_to(data_dir)}: {size_mb:.2f} MB")

# Calculate total size
total_csv_mb = sum(f.stat().st_size for f in csv_files) / (1024 * 1024)
total_parquet_mb = sum(pf.stat().st_size for pf in parquet_files) / (1024 * 1024)
if total_parquet_mb > 0:
    overall_compression = total_csv_mb / total_parquet_mb
    print(f"\n{'=' * 80}")
    print(f"Overall: {total_csv_mb:.2f} MB → {total_parquet_mb:.2f} MB ({overall_compression:.2f}x smaller)")
    print(f"Space saved: {total_csv_mb - total_parquet_mb:.2f} MB")
else:
    print(f"\n{'=' * 80}")
    print("No parquet files found yet.")
print(f"{'=' * 80}")
````
