```python
# read the ride year 2024 parquet files into a pandas dataframe
from glob import glob


def read_parquet_files(path):
    """Reads multiple parquet files into a single pandas DataFrame."""
    df = pd.concat([pd.read_parquet(f) for f in glob.glob(path)], ignore_index=True)
    return df
```
